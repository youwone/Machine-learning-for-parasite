```python
from pdb import set_trace
from pandas import Series, DataFrame
import pandas as pd

import re


def DIAGNOSIS():
    file = open(r'Diagnosis_term.txt', 'r')

    lDIagnosis = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        lDIagnosis.append(sRow)
    file.close()
    return lDIagnosis


def OS():
    import os
    lFiles = os.listdir(r'./TextMining_result')
    return lFiles


def GENDER():
    file = open(r'Gender.txt', 'r')

    lGender = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        lGender.append(sRow)
    file.close()
    return lGender


def OLD():
    file = open(r'Old.txt', 'r')

    lOld = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        lOld.append(sRow)
    file.close()
    return lOld


def NATION():
    file = open(r'Nation_list_20180125.txt', 'r')

    lNation = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        lNation.append(sRow)
    file.close()
    return lNation


def SYMPTOM():
    file = open(r'medicalsymptom_wiki.txt', 'r')

    lSymptom = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        sRow = sRow.replace('\t', '').upper()
        lSymptom.append(sRow)
    file.close()
    return lSymptom

def SITE():
    file = open(r'medicalsymptom_wiki_site.txt', 'r')

    lSite = []
    for sRow in file:
        sRow = sRow.replace('\n', '').upper()
        sRow = sRow.replace('\t', '').upper()
        lSite.append(sRow)
    file.close()
    return lSite

def fParse_PUBMED():

    lDiagnosis = DIAGNOSIS()
    lGender = GENDER()
    lOld = OLD()
    lNation = NATION()
    lSymptom = SYMPTOM()
    lSite = SITE()
    Age_pattern = re.compile(r"\d+.?year.?.old")
    Age_extract = re.compile(r"\d+")
    abstract=[]
    for sFile in OS():
        sParasite = sFile.replace('(', '').replace(')', '').split(' ')[0]

        dRESULT = {}
        dRESULT[sParasite] = {}
        with open(r'./TextMining_result/%s' % sFile, encoding='UTF8') as Input:
            for sRow in Input:
                lCol = sRow.replace('\n', '').upper()

                if sRow[:4] =='PMID':
                    PMID = sRow
                    dRESULT[sParasite][PMID] = {}


                if sRow[:8] == 'Abstract':

                    lCol = sRow.replace('\n', '').upper()
                    for sDiagnosis in lDiagnosis:

                        if ' '+ sDiagnosis + ' ' in lCol or ' '+sDiagnosis + ',' in lCol or ' '+ sDiagnosis+ '.' in lCol or ' '+ sDiagnosis+ '.\n' in lCol or ' '+ sDiagnosis + ':' in lCol or ' '+sDiagnosis + ')' in lCol or ' '+sDiagnosis + '),' in lCol or ' '+sDiagnosis+ '(' in lCol or '(' + sDiagnosis in lCol or '(' + sDiagnosis + ',' in lCol or '(' + sDiagnosis + ')' in lCol:
                            try:
                                dRESULT[sParasite][PMID]['Diagnosis'] = sDiagnosis

                            except ValueError:
                                dRESULT[sParasite][PMID]['Diagnosis'] = 0

                    for sGender in lGender:

                        if ' '+ sGender + ' ' in lCol or ' '+sGender + ',' in lCol or ' '+ sGender+ '.' in lCol or ' '+ sGender+ '.\n' in lCol or ' '+ sGender + ':' in lCol or ' '+sGender + ')' in lCol or ' '+sGender + '),' in lCol or ' '+sGender+ '(' in lCol or '(' + sGender in lCol or '(' + sGender + ',' in lCol or '(' + sGender + ')' in lCol:
                            try:
                                dRESULT[sParasite][PMID]['GENDER'] = sGender

                            except ValueError:
                                dRESULT[sParasite][PMID]['GENDER'] = 0

                    for sOld in lNation:

                        if ' '+ sOld + ' ' in lCol or ' '+sOld + ',' in lCol or ' '+ sOld+ '.' in lCol or ' '+ sOld+ '.\n' in lCol or ' '+ sOld + ':' in lCol or ' '+sOld + ')' in lCol or ' '+sOld + '),' in lCol or ' '+sOld+ '(' in lCol or '(' +sOld in lCol or '(' + sOld + ',' in lCol or '(' + sOld + ')' in lCol:
                            try:
                                dRESULT[sParasite][PMID]['NATION'] = sOld

                            except ValueError:
                                dRESULT[sParasite][PMID]['NATION'] = 0
                    for sSymptom in lSymptom:

                        if ' '+ sSymptom + ' ' in lCol or ' '+sSymptom + ',' in lCol or ' '+ sSymptom+ '.' in lCol or ' '+ sSymptom+ '.\n' in lCol or ' '+ sSymptom + ':' in lCol or ' '+sSymptom + ')' in lCol or ' '+sSymptom + '),' in lCol or ' '+sSymptom+ '(' in lCol or '(' +sSymptom in lCol or '(' + sSymptom + ',' in lCol or '(' + sSymptom + ')' in lCol:
                            try:
                                dRESULT[sParasite][PMID]['SYMPTOM'] = sSymptom

                            except ValueError:
                                dRESULT[sParasite][PMID]['SYMPTOM'] = 0
                    for sSite in lSite:

                        if ' '+ sSite + ' ' in lCol or ' '+sSite + ',' in lCol or ' '+ sSite + '.' in lCol or ' '+ sSite + '.\n' in lCol or ' '+ sSite + ':' in lCol or ' '+ sSite + ')' in lCol or ' '+ sSite + '),' in lCol or ' '+sSite+ '(' in lCol or '(' +sSite in lCol or '(' + sSite + ',' in lCol or '(' + sSite + ')' in lCol:
                            try:
                                dRESULT[sParasite][PMID]['SITE'] = sSite

                            except ValueError:
                                dRESULT[sParasite][PMID]['SITE'] = 0

                    for iAge in set([Age_extract.findall(Age)[0] for Age in Age_pattern.findall(sRow, re.IGNORECASE)]):
                        try:

                            dRESULT[sParasite][PMID]['OLD'] = iAge

                        except ValueError:
                            dRESULT[sParasite][PMID]['OLD'] = 0


        Result=pd.DataFrame(index=['Diagnosis','GENDER','OLD','NATION','SYMPTOM','SITE'])

        for Keys in dRESULT.keys():

            Loop = dRESULT.get(Keys)

            Result=pd.DataFrame(Loop)

            Result.to_csv(r"./Database_\%s.csv" % Keys, mode="w")

```
